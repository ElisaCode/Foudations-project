```python
import bs4
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import requests
from urllib.request import Request, urlopen
import unicodedata
import re 
```

```python
def get_soup(url):
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3)'
                                             ' AppleWebKit/537.36 (KHTML, like Gecko)'
                                             ' Chrome/35.0.1916.47 Safari/537.36'})
    web_byte = urlopen(req).read()
    soup = BeautifulSoup(web_byte, 'html.parser')
    return soup
url = "https://www.conjugacao.com.br/verbos-populares/"
html_object = get_soup(url)
```

```python
main_url="https://www.conjugacao.com.br/verbos-populares/"
def make_all_links(main_url):
    links_list = []
    links_list.append(main_url)
    r_list = list(range(2,51))
    for n in r_list:
        link = main_url + str(n) + '/'
        links_list.append(link)
    print(links_list)
    return links_list
```

```python
all_pages = make_all_links(main_url)
```

```python
def get_all_pages(page_list):
    all_verbs = []
    for page in page_list: ###### page_list não deveria estar definida em algum lugar?
        print(page)
        page_soup = get_soup(page)
        lista_verbos = page_soup.find(class_='wrapper').findAll('li')
        for tag in lista_verbos:
            clean_tag = tag.find('a')
            link = "https://www.conjugacao.com.br" + clean_tag["href"]
            verb = clean_tag.text
            pair = (verb, link)
            all_verbs.append(pair)
    print(all_verbs)
    return all_verbs
```

```python
lista_big = get_all_pages(all_pages) #links all pages 
```

```python
list_links_verbs = []
for link in lista_big:
    vlink = link[1]
    list_links_verbs.append(vlink)
    
    
```
